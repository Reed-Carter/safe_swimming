```python
import pandas as pd
```

```python
station_number_df = pd.read_csv('trash.csv', dtype=str)
station_numbers = station_number_df['StationNumber'].tolist()
print(station_numbers)
len(station_numbers)
# station_numbers[144]
```

```python
# URL to scrape
url = 'https://waterdata.usgs.gov/monitoring-location/01208837/#parameterCode=00010&period=P7D'
# url = 'https://waterdata.usgs.gov/monitoring-location/0329247350/#parameterCode=00010&period=P7D'

# Read tables from the URL
tables = pd.read_html(url)

# Select the table of interest (you may need to adjust the index if there are multiple tables)
df = tables[1]  # Assuming the desired table is at index 0

# Select rows with specific values in 'Metadata Element' column
df = df[df['Metadata Element'].isin(['Site identification number', 'Decimal latitude', 'Decimal longitude'])]
df = df.drop('Metadata Code', axis=1)

# Reset the index of the DataFrame
df = df.reset_index(drop=True)
df = df.rename(columns={'Metadata Element': 'index'})

df = df.pivot(columns='index', values='Location Metadata')

# Rename the columns
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'Site_identification_number': 'StationNumber'})

# Merge rows into one row
df = df.ffill().dropna()

df = df.reset_index(drop=True)

# Print the DataFrame
display(df)
```

```python
df_coordinates = pd.DataFrame()
```

```python
  

for index, value in enumerate(station_numbers):
     if index > 4475:
        # URL to scrape
        url = f'https://waterdata.usgs.gov/monitoring-location/{value}/#parameterCode=00010&period=P7D'

        # Read tables from the URL
        tables = pd.read_html(url)

        # Select the table of interest (you may need to adjust the index if there are multiple tables)
        df = tables[1]  # Assuming the desired table is at index 1

        # Select rows with specific values in 'Metadata Element' column
        df = df[df['Metadata Element'].isin(['Site identification number', 'Decimal latitude', 'Decimal longitude'])]
        df = df.drop('Metadata Code', axis=1)

        # Reset the index of the DataFrame
        df = df.reset_index(drop=True)
        df = df.rename(columns={'Metadata Element': 'index'})

        df = df.pivot(columns='index', values='Location Metadata')

        # Rename the columns
        df.columns = df.columns.str.replace(' ', '_')
        df = df.rename(columns={'Site_identification_number': 'StationNumber'})

        # Merge rows into one row
        df = df.ffill().dropna()
        df = df.reset_index(drop=True)

        df_coordinates = pd.concat([df_coordinates, df], ignore_index=True)

        print(f'added {index+1} stations number to df_coordinates')


# # Print the DataFrame
display(df_coordinates)
```
